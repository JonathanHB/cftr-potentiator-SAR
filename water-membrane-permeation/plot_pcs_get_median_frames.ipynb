```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
npy_files = [f for f in os.listdir() if f[-4:] == ".npy"]

for f in npy_files:
    pc = np.load(f)
    plt.plot(pc, label = f)

plt.legend()
print(npy_files)
```

```python
nbins = 51
binbounds = np.arange(0,nbins,1)

test_bins = [1,10,20,30,40]

median_frames = np.zeros((len(npy_files), len(test_bins)))

#loop over trjs
for i, f in enumerate(npy_files):
    print(f)
    pcs = np.load(f)
    bins = np.digitize(pcs, binbounds)

    for j, bin in enumerate([1,10,20,30,40]):
        frames = np.where(bins==bin)[0]
        plt.hist(frames, range = [0,2000], alpha = 0.2)
        #this equals the median for odd number of frames and is the larger of the two middle frames for odd numbers
        median_frames[i][j] = sorted(frames)[len(frames)//2] #np.median(frames)
        #print(frames)
        #print(approx_median_inlist(frames[0]))
    plt.show()

print(median_frames)
```
